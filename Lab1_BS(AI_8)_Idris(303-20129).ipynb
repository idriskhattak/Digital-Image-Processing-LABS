#   Importing modules

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

#  1) Loading the image

```python
picture = Image.open('greyscale-009.webp')
```

```python
picture
```

# 2) Printing its basic properties such as width, height, and number of channels

To see image properties first we have to convert it to a numpy array

```python
type(picture)
```

Converting it to an array

```python
pic_arr = np.asarray(picture)
```

The is the width, height and channel of the image

```python
pic_arr.shape
```

# 3) Calculating and displaying its basic statistics such as mean, median, and standard deviation of pixel values

```python
print('The mean of the image is :',np.mean(pic_arr))
print('The medina of the image is :',np.median(pic_arr))
print('The median of the image is :',np.std(pic_arr))
```
